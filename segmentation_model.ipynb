# Automotive Market Demand & Customer Segmentation Analysis

This notebook performs EDA, Customer Segmentation using K-Means, and Demand Forecasting.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
# Load Data
df = pd.read_csv('data.csv')
df.head()
```

## 1. Exploratory Data Analysis (EDA)

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['annual_income'], bins=30, kde=True)
plt.title('Income Distribution')
plt.show()
```

## 2. Customer Segmentation (K-Means)

```python
features = ['age', 'annual_income', 'dealer_visit_count', 'purchase_probability']
scaler = StandardScaler()
scaled_features = scaler.fit_transform(df[features])

kmeans = KMeans(n_clusters=3, random_state=42)
df['cluster'] = kmeans.fit_predict(scaled_features)

sns.scatterplot(data=df, x='annual_income', y='purchase_probability', hue='cluster', palette='deep')
plt.title('Customer Segments')
plt.show()
```
